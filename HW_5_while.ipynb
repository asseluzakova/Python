1. Рисование прямоугольника из звездочек:

Задача: Напишите программу, которая рисует прямоугольник из звездочек (*). Пользователь вводит высоту и ширину прямоугольника (по очереди примите два значения от пользователя, не одновременно).

Инструменты: Используйте внешний цикл for для перебора строк и внутренний цикл for для перебора столбцов в каждой строке. Используйте print

Вывод:

 - Предположим, пользователь вводит высоту 3 и ширину 5. Тогда вывод программы будет следующим:

     *****

     *****

     *****


```python
number1=int(input("высота:"))
number2=int(input("ширина:"))
for i in range(number1):
    for j in range(number2):
        print("*", end='')
    print('\n')
```

Задача: Напишите программу, которая для чисел от 1 до N (N задается пользователем) выводит их степени от 1 до M (M также задается пользователем).

Инструменты: Используйте один цикл for для перебора чисел от 1 до N и вложенный цикл for для возведения каждого числа в степени от 1 до M.

Вывод:

- Если пользователь задает N = 3 и M = 4, то вывод будет следующим (выводятся степени каждого числа от 1 до 4):

     1^1 = 1, 1^2 = 1, 1^3 = 1, 1^4 = 1



     2^1 = 2, 2^2 = 4, 2^3 = 8, 2^4 = 16



     3^1 = 3, 3^2 = 9, 3^3 = 27, 3^4 = 81

```python
N=int(input("N="))
M=int(input("M="))
for i in range(1,N+1):
    for j in range(1,M+1):
        print(i,'^',j,'=',i**j,', ', end='')
    print('\n')
```

1. Создание простого счетчика времени:

Задача: Напишите программу, которая ведет обратный отсчет от заданного пользователем времени в секундах до нуля.

Инструменты: Используйте цикл while для уменьшения счетчика каждую секунду и вывода оставшегося времени. Программа должна остановиться, когда счетчик достигнет нуля.

Вывод:

- Если пользователь задает время 5 секунд, то вывод будет выглядеть так:

     5



     4



     3



     2



     1



     Время истекло!

```python
number=int(input("введите число:"))
count=1
while number>=count:
    print(number)
    number-=1
else:
    print("Время истекло!")
```

2.Программа угадывания числа:

Задача: Напишите программу, где компьютер загадывает число от 1 до 100, а пользователь пытается его угадать. После каждой попытки программа должна сообщать, больше или меньше загаданное число, чем введенное пользователем. Процесс продолжается до тех пор, пока число не будет угадано.

Инструменты: Используйте цикл while для повторения действий, пока пользователь не угадает число.

Вывод:   

- Представим, что компьютер загадал число 42, а пользователь вводит следующие числа: 50, 30, 45, 42. Тогда вывод будет таким:

     Ваше число: 50

     Загаданное число меньше!

     Ваше число: 30

     Загаданное число больше!

     Ваше число: 45

     Загаданное число меньше!

     Ваше число: 42

     Вы угадали число!

```python
from random import randint
N=randint(1,100)
while K!=N:
    K=int(input("Ваше число:"))
    if K<N:
        print("Загаданное число меньше!")
    elif K>N:
        print("Загаданное число больше!")
    else:
        print("Вы угадали число!")
print(K)
```
